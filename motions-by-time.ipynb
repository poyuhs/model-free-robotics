```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataset
dataset_df = pd.read_csv("D://_NYCU GIA//_Thesis//_gh//_data//yawAngByTime_n2.csv")

# Extract the time data (assuming time is the index)
iteration = dataset_df.index

# Plotting the data
plt.figure(figsize=(10, 6))

# Iterate through each column and plot
for column in dataset_df.columns[::-1]:
    plt.plot(iteration, dataset_df[column], label=column)

# Adding labels and title
plt.xlabel('Iteration')
plt.ylabel('Angle (degree)')
plt.grid(True)
plt.xlim(0, 200)
plt.ylim(0, 100)
plt.title('Robot motions by varying the α values with β=0 (N=2)')
plt.legend(title='α values', loc='upper right')

# Display the plot
plt.show()
```

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import datasets for N=1, 2, 3, 4
df_n1 = pd.read_csv("D://_NYCU GIA//_Thesis//_gh//_data//yawAngByTime_n1.csv")
df_n2 = pd.read_csv("D://_NYCU GIA//_Thesis//_gh//_data//yawAngByTime_n2.csv")
df_n3 = pd.read_csv("D://_NYCU GIA//_Thesis//_gh//_data//yawAngByTime_n3.csv")
df_n4 = pd.read_csv("D://_NYCU GIA//_Thesis//_gh//_data//yawAngByTime_n4.csv")

# Desired angles (assuming 10, 20, ..., 90)
desired_angles = np.arange(10, 100, 10)

# Calculate deviations
deviation_n1 = np.abs(desired_angles - df_n1.iloc[-1].values)
deviation_n2 = np.abs(desired_angles - df_n2.iloc[-1].values)
deviation_n3 = np.abs(desired_angles - df_n3.iloc[-1].values)
deviation_n4 = np.abs(desired_angles - df_n4.iloc[-1].values)

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(desired_angles, deviation_n1, marker='o', label='N=1')
plt.plot(desired_angles, deviation_n2, marker='o', label='N=2')
plt.plot(desired_angles, deviation_n3, marker='o', label='N=3')
plt.plot(desired_angles, deviation_n4, marker='o', label='N=4')

# Adding labels and title
plt.xlabel('α value (degree)')
plt.ylabel('Deviation value (degree)')
plt.grid(True)
plt.title('Deviation values for each N with β=0')
plt.legend()

# Display the plot
plt.show()
```
